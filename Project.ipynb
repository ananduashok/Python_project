(1 mark) Analysis Tasks: 

1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees. (2 marks)
   

```python
# calculate the team-wise distribution

team_distribution = df['Team'].value_counts().reset_index()
team_distribution.columns = ['Team','Count']

# calculate percentage split

employee_len = len(df)
team_distribution['Percentage'] = (team_distribution['Count'] / employee_len) * 100
team_distribution
```

```python
# Display it visually

plt.figure(figsize=(20,6))

# Normalize the colormap to the number of bars
colors = plt.get_cmap('summer', len(df))(range(len(df)))

bars = plt.bar(
    team_distribution['Team'],
    team_distribution['Count'],
    color = colors
)

# Add percentage labels on top of the bars
for bar, percent in zip(bars, team_distribution['Percentage']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 2, f'{percent:.1f}%', ha='center', color='white', fontsize=10)

# Customize the plot
plt.title('Employee Distribution by Team (Percentage)', fontsize=16)
plt.xlabel('Team', fontsize=14)
plt.ylabel('Percentage (%)', fontsize=14)
plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

2. Segregate employees based on their positions within the company. (2 marks) 

```python
df.head()
```

```python
position_distribution = df['Position'].value_counts().reset_index()
position_distribution.columns = ['Position','Count']
position_distribution
```

```python
# Visual representation

plt.figure(figsize = (10,6))
explode = [0.1 if i == position_distribution['Count'].max() else 0.05 for i in position_distribution['Count']]  # Highlight the largest segment

plt.pie(
    position_distribution['Count'],
    labels = position_distribution['Position'],
    autopct='%1.1f%%',
    startangle=10,
    explode=explode,
    shadow = True
)

plt.title('Employee Segregation by Position', fontsize = 15, font = 'Calibri')
plt.tight_layout()
plt.legend(fontsize=8)

plt.show()
```

3. Identify the predominant age group among employees. (2 marks) 

```python
df.head()

# we can use the mode function to get the predominant age group
dominant_age = df.Age.mode()
print("The predominant age group in this data with the most employees is: ",dominant_age[0])
```

```python
age_group = df['Age'].value_counts().reset_index()
age_group.columns=['Age','Count']
age_group['Percentage'] = (age_group['Count']/len(df))*100
age_group
```

```python
# display it visually

plt.figure(figsize=(20,10))
explode = [0.2 if i == age_group['Count'].max() else 0 for i in age_group['Count']]  # Highlight the largest segment

plt.pie(
    age_group['Count'],
    labels = age_group['Age'],
    autopct='%1.1f%%',
    explode = explode
)

plt.title("Predominant Age Group among the Employees")
plt.tight_layout()

plt.show()
```

4. Discover which team and position have the highest salary expenditure. (2 marks) 

```python
# total salary expenditure by team and position

team_salary = df.groupby('Team')['Salary'].sum().reset_index()
position_salary = df.groupby('Position')['Salary'].sum().reset_index()
print(team_salary)
position_salary
```

```python
# team with highest salary expenditure

highest_team = team_salary.loc[team_salary['Salary'].idxmax()]
print(highest_team)
```

```python
# position with high salary expenditure

highest_position = position_salary.loc[position_salary['Salary'].idxmax()]
highest_position
```

```python
# visualize

#team salary expenditure

plt.figure(figsize = (10,6))
bars = plt.bar(
    team_salary['Team'],
    team_salary['Salary'],
    color='skyblue',
    edgecolor='black'
)

# Highlight the bar with the highest salary and annotate it
max_team_bar = bars[team_salary['Salary'].idxmax()]
plt.text(max_team_bar.get_x() + max_team_bar.get_width() / 2, max_team_bar.get_height() + 2000,
         f'${highest_team["Salary"]:,.0f}', ha='center', fontsize=12, color='black')

plt.title('Team-wise Salary Expenditure')
plt.xlabel('Team')
plt.ylabel('Salary')
plt.xticks(fontsize=10,rotation=80)
plt.yticks(fontsize=10)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
# visualize Position wise salary expenditure

plt.figure(figsize = (10,6))

bars = plt.bar(
    position_salary['Position'],
    position_salary['Salary'],
    color='lightgreen',
    edgecolor='black'
)

max_position_bar = bars[position_salary['Salary'].idxmax()]
plt.text(max_position_bar.get_x() + max_position_bar.get_width() / 2, max_position_bar.get_height() + 2000,
         f'${highest_position["Salary"]:,.0f}', ha='center', fontsize=10, color='black')

plt.title('Position-wise Salary Expenditure')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.tight_layout()

plt.show()
```

5. Investigate if there's any correlation between age and salary, and represent it visually. 

```python
# correlation between age and salary

correlation = df['Age'].corr(df['Salary'])
print(f"Correlation between Age and Salary: {correlation:.2f}")
```

The corr() method calculates the Pearson correlation coefficient between age and salary. A positive correlation means that as age increases, salary tends to increase (and vice versa). Since the correlation is close to 0, we have a positive correaltion but not so strong.

```python
#visualizing the correlation 

plt.figure(figsize =(10,6))

sns.scatterplot(
    x='Age',
    y='Salary',
    data = df
)

# adding a regression line to show the trend

sns.regplot(
    x='Age',
    y='Salary',
    data=df,
    color='red',
    scatter=False
)

# Customize plot
plt.title('Correlation between Age and Salary', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Salary ($)', fontsize=14)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
